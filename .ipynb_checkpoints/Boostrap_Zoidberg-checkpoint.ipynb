```python
import numpy as np 
import pandas as pd 
import random as rn

import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import graphviz
import pydotplus

from IPython.display import Image 
from io import StringIO

from sklearn import tree
from sklearn.tree import DecisionTreeClassifier, export_graphviz
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn import datasets, metrics, preprocessing
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.metrics import classification_report
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, MinMaxScaler

from sklearn.model_selection import cross_val_score

from sklearn.model_selection import train_test_split

from keras.datasets import mnist
```

```python
class MNISTData:
    def __init__(self):
        # Load the MNIST dataset
        (self.train_X, self.train_y), (self.test_X, self.test_y) = mnist.load_data()
        
        # Preprocess the data
        #self.train_X = train_X.reshape((60000, 28 * 28)).astype('float32') / 255
        #self.train_y = train_y.astype('float32')
        #self.test_X = test_X.reshape((10000, 28 * 28)).astype('float32') / 255
        #self.test_y = test_y.astype('float32')
    
    def get_reshaped_train_data(self):
        self.train_X = self.train_X.reshape(-1, 28, 28, 1).astype('float32').reshape(-1,28*28)
        self.train_y = self.train_y.astype('float32')
        return self.train_X, self.train_y
    
    def get_reshaped_train_data(self):
        self.test_X = self.test_X.reshape(-1, 28, 28, 1).astype('float32').reshape(-1,28*28)
        self.test_y = self.test_y.astype('float32')
        
        return self.test_X, self.test_y

    def get_train_data(self):
        return self.train_X, self.train_y

    def get_test_data(self):
        return self.test_X, self.test_y
    
    def show_the_ditributions_of_each_digit(self):
            counts = np.bincount(self.train_y)
            print("counts : ", counts)

            fig, ax = plt.subplots() #tuple figure et axes
            ax.bar(range(10), counts)
            ax.set_xlabel('Digit')
            ax.set_ylabel('Number of images')
            ax.set_title('Distributions of each digit')

            for i, count in enumerate(counts):
                ax.text(i, count+50, str(count), ha='center', fontsize=10)

            plt.xticks(range(10))  # Définir les abscisses de 0 à 9
            plt.show()
        
    def display_image(self, dataset, place):
        """
        On montre une image spécifique depuis un dataset choisi
        
        Args :
        dataset: correspond à soit "test" ou "train"
        place: correspond au num de l'image 
        """
        if dataset == "train":
            image = self.train_X[place]
            label = self.train_y[place]
        elif dataset == "test":
            image = self.test_X[place]
            label = self.test_y[place]
        else :
            print("Dataset erreur")
            return

        plt.imshow(image, cmap='gray') #cmap, colormap => niveaux de gris 
        plt.title(f"Label: {label}")
        plt.show()
        
    def see_some_data_samples(self):
        fig = plt.figure(figsize=(10,7))
        for i in range(10):
            ax = fig.add_subplot(3, 5, i+1)
            ax.imshow(self.train_X[i], cmap='gray')
            ax.set_title('Label : {}'.format(self.train_y[i]))
            plt.axis('off')
    
    def digit_occurrences(self):
        counts = np.bincount(self.train_y)
        occurrences = counts / len(self.train_y)
        digit_occurrences = [(i, occurrences[i]) for i in range(10)]
        return digit_occurrences
```

```python
mnist_dt = MNISTData()
mnist_dt.see_some_data_samples()
```

# Prédictions

```python
X_train, y_train = mnist_dt.get_train_data()
X_test, y_test = mnist_dt.get_test_data()

training_dt_x, training_dt_y = mnist_dt.get_reshaped_train_data()
test_dt_x, test_dt_y = mnist_dt.get_reshaped_train_data()
```

# DECISION TREE

### Model Decision Tree
https://scikit-learn.org/stable/modules/tree.html#classification

```python
clf = tree.DecisionTreeClassifier()
%time clf.fit(training_dt_x, training_dt_y)
```

```python
training_pred_dt = clf.predict(training_dt_x)
test_pred_dt = clf.predict(test_dt_x)
```

### Accuracy for our baseline

```python
%time clf.score(test_dt_x, test_dt_y)
```

```python
train_acc = (training_pred_dt == training_dt_y).sum() / len(training_dt_y)
print("Training accuracy : {}".format(train_acc))
```

```python
test_acc = (test_pred_dt == test_dt_y).sum() / len(test_dt_y)
print("Test accuracy : {}".format(test_acc))
```

### Visualisation

```python
dot_data = StringIO()
export_graphviz(clf, out_file=dot_data,max_depth=3,
                filled=True, rounded=True,  special_characters=True)  
```

```python
graph =pydotplus.graph_from_dot_data(dot_data.getvalue()) 
Image(graph.create_png())
```

# SVM

https://scikit-learn.org/stable/modules/generated/sklearn.svm.SVC.html#sklearn.svm.SVC

https://scikit-learn.org/stable/modules/svm.html#classification

```python
training_dt_x =training_dt_x[0:10000,:]
training_dt_y =training_dt_y[0:10000]
test_dt_x =test_dt_x[0:2000,:]
test_dt_y =test_dt_y[0:2000]
```

```python
pipe_1 = Pipeline([('scaler', MinMaxScaler()), ("classifier", SVC(kernel = 'linear', C = 1))])
%time pipe_1.fit(training_dt_x, training_dt_y.ravel())
```

```python
acc = cross_val_score(pipe_1, training_dt_x, training_dt_y.ravel(), cv=2)
print("Training Accuracy : {}".format(acc))
print("Training Accuracy Mean : {}".format(acc.mean()))
print("Training Accuracy in % : {:.2f} %".format(acc.mean()*100))
```

```python
pipe_1_predict_result = pipe_1.predict(test_dt_x)
```

```python
cm = confusion_matrix(test_dt_y, pipe_1_predict_result)
cfmd = ConfusionMatrixDisplay(confusion_matrix=cm)
cfmd.plot()
plt.title('Confusion Matrix')
plt.show()
```

```python
CR = classification_report(test_dt_y, pipe_1_predict_result)
print('Classification report \n')
print(CR)
```
